### Now let's see how df_new = df.copy() differs from df_new = df ###



```python
df_original = dataset['train'].to_pandas()

df_altered = df_original
```

Now let's make some changes to the altered dataframe

```python
median_sal = df_altered['salary_year_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_sal)

df_altered.loc[:5, 'salary_year_avg']
```

```python
df_original['salary_year_avg']
```

As we can see the changes made to the altered dataframe are also reflected to the original dataframe
TO avoid this we use .copy() 

```python
df_altered = df_original.copy()

median_sal = df_altered['salary_year_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(
    median_sal)

df_altered.loc[:5, 'salary_year_avg']
```
